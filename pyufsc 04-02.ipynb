```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def quadrado(x):
    return x**2
```

```python
def cubo(x):
    return x**3
```

```python
def baskara(a, b, c):
    x1 = (-b+np.sqrt(b**2-4*a*c))/(2*a)
    x2 = (-b-np.sqrt(b**2-4*a*c))/(2*a)
    return (x1, x2)
```

## Teste das Funções

```python
assert quadrado(2) == 4
assert quadrado(3) == 9
assert quadrado(5) == 25
assert quadrado(1) == 1
assert quadrado(12) == 144
```

```python
assert cubo(3) == 27
assert cubo(1) == 1
assert cubo(2) == 8
assert cubo(5) == 125
assert cubo(4) == 64
```

```python
assert baskara(2, -5, -3) == (3, -0.5)
assert baskara(2, -5, 3) == (1.5, 1)
assert baskara(1,0, -4) == (2, -2)
assert baskara(1, -4, -9) == (2+np.sqrt(13), 2-np.sqrt(13))
assert baskara(-1, 0, 9) == (-3, 3)
```
